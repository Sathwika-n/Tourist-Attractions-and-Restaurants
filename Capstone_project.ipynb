```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

URL = "https://en.wikipedia.org/wiki/Tourist_attractions_in_the_United_States"

# Fetch the content of the URL
response = requests.get(URL)
soup = BeautifulSoup(response.content, 'html.parser')

# Locate the table for top tourist attractions
table = soup.find_all('table')[0] # Based on the structure, assuming the first table contains the required information

# Extracting the data from the table
places = []
for row in table.find_all('tr')[1:]:  # Skip the header row
    columns = row.find_all('td')
    place = columns[0].get_text(strip=True)
    location = columns[1].get_text(strip=True)
    visitors = columns[2].get_text(strip=True)
    places.append({
        "Place": place,
        "Location": location,
        "Visitors (millions)": visitors
    })

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(places)

# Print the DataFrame
df

# Optionally, save the DataFrame to a CSV file
# df.to_csv("tourist_attractions.csv", index=False)
```

```python
# Remove bracketed values from 'Visitors (millions)' column
df['Visitors (millions)'] = df['Visitors (millions)'].str.replace(r'\[.*\]', '', regex=True)

# Print the DataFrame

df
```

```python
url = "https://www.buzzfeed.com/hannahloewentheil/what-to-visit-in-every-state"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# Find all relevant tags (this might need to be adjusted based on the actual HTML structure)
entries = soup.find_all('h2')  # Assuming the data is in h2 tags, adjust accordingly

states = []
destinations = []

for entry in entries:
    text = entry.get_text()
    # Assuming the format is "State: Destination"
    parts = text.split(":")
    if len(parts) == 2:
        state, destination = parts
        states.append(state.strip())
        destinations.append(destination.strip())

# Create a DataFrame
State_TopDestination = pd.DataFrame({
    'State': states,
    'Top Destination': destinations
})

State_TopDestination
```

```python

# Remove newline characters and numbers from 'State' column
State_TopDestination['State'] = State_TopDestination['State'].str.replace('\n', '').str.extract('(\D+)', expand=False).str.strip()

State_TopDestination
```

```python

# Remove the period at the beginning of each state name
State_TopDestination['State'] = State_TopDestination['State'].str.lstrip('.')

State_TopDestination
```
